```python
import numpy as np
import matplotlib.pyplot as plt

def rectangular_pulse(time):
    # Pulse width of 1 centered at zero
    return np.where(np.abs(time) <= 0.5, 1, 0)

time = np.linspace(-2, 2, 1000)
pulse = rectangular_pulse(time)

plt.figure(figsize=(10, 6))
plt.plot(time, pulse, color='darkorange', linewidth=2.5, label='Rectangular Pulse')
plt.title('Rectangular Pulse Function Π(t)', fontsize=16)
plt.xlabel('Time (t)', fontsize=14)
plt.ylabel('Amplitude Π(t)', fontsize=14)
plt.grid(which='both', linestyle='--', alpha=0.7)
plt.ylim(-0.2, 1.2)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rect(t):
    return np.where(np.abs(t) <= 0.5, 1, 0)

t = np.linspace(-5, 5, 1000)

plt.figure(figsize=(15, 10))

# Plot x(3t - 0.5)
plt.subplot(3, 1, 1)
x1 = rect(3 * t - 0.5)
plt.plot(t, x1, 'b-', linewidth=2)
plt.title('x(3t - 0.5)', fontsize=14)
plt.grid(True)
plt.ylim(-0.5, 1.5)
plt.xlabel('t')
plt.ylabel('Amplitude')

# Plot x(0.5t + 3)
plt.subplot(3, 1, 2)
x2 = rect(0.5 * t + 3)
plt.plot(t, x2, 'r-', linewidth=2)
plt.title('x(0.5t + 3)', fontsize=14)
plt.grid(True)
plt.ylim(-0.5, 1.5)
plt.xlabel('t')
plt.ylabel('Amplitude')

# Plot x(t - 2) - t
plt.subplot(3, 1, 3)
x3 = rect(t - 2) - t
plt.plot(t, x3, 'g-', linewidth=2)
plt.title('x(t - 2) - t', fontsize=14)
plt.grid(True)
plt.xlabel('t')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
